```python
# this dataset is the data that were used to compare the results of the experiments
data = pandas.read_csv('../data/geolife/compatible.csv')
print(f'shape: {data.shape}')
print(f'#unique agents: {len(data['AgentID'].unique())}')   

print(f'columns: {data.columns}')

```

```python
background = '../data/geolife/map.png'
#['AgentID', 'ArrivingTime', 'Latitude', 'Longitude', 'Distance']
# plot the data
plt.figure(figsize=(5, 5))
plt.scatter(data['Longitude'], data['Latitude'], s=1, alpha=0.1)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geolife Dataset')
plt.imshow(plt.imread(background), extent=[116.165, 116.628, 39.748, 40.038])
plt.show()
```
